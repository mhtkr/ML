```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Sample data (Hours studied vs Pass/Fail)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (Hours studied)
Y = np.array([0, 0, 0, 1, 1])                # Dependent variable (Pass/Fail)

# Creating and training the model
model = LogisticRegression()
model.fit(X, Y)

# Making predictions
Y_pred = model.predict(X)

# Printing the model's parameters and accuracy
print(f"Intercept (c): {model.intercept_[0]}")
print(f"Slope (m): {model.coef_[0][0]}")
print(f"Predicted Classes: {Y_pred}")
print(f"Accuracy: {accuracy_score(Y, Y_pred)}")

# Confusion matrix
conf_matrix = confusion_matrix(Y, Y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")

# Plotting sigmoid function
X_test = np.linspace(0, 6, 300).reshape(-1, 1)  # Test range for hours studied
probabilities = model.predict_proba(X_test)[:, 1]  # Probability of passing

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X_test, probabilities, color='red', label='Sigmoid Curve')
plt.axhline(0.5, color='green', linestyle='--', label='Threshold (0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression')
plt.legend()
plt.show()
```
